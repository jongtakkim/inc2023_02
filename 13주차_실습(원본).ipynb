###1. 아래 리스트의 값을 사용하여 순위를 구하시오
###   score=[70,80,60,100,50]

```python
score=[70,80,60,100,50]
rank=[0]*5
n=len(score)
for i in range(n) :
  rank[i] = 1
  for j in range(n) :
    if score[i] < score[j] :
      rank[i] += 1

for i in range(n) :
  print(f"점수 : {score[i]}점  등수 : {rank[i]}등")
```

###2. 아래 리스트를 사용하여 80점 이상이면 ‘합격’ 80점 미만이면 ‘불합격을 출력하고 합격한 학생의 수를 출력하시오
###    score=[60,80,90,70,95]

```python
score=[60,80,90,70,95]
cnt = 0
for i in range(len(score)) :
  if score[i] > 80 :
    result='합격'
    cnt += 1
  else :
    result = '불합격'
  print(f"{i+1}번 학생은 {result}입니다.")
print(f'합격한 학생의 수 : {cnt}')
```

###3. 수강과목수와 각각의 수강 과목점수를 입력받아 리스트에 저장하고 80점 이상인 과목은 “합격", 80점 미만은 ”불합격“이라고 한다. 합격" 과 "불합격"한 과목이 수를 <출력결과>와 같이 출력하시오


```python
score=[]
p,f=0,0
cnt=int(input('수강 과목 수 : '))

for i in range(cnt) :
  s = int(input(f'{i+1}번째 점수 : '))
  score.append(s)

for i in range(cnt) :
  if score[i] > 80 :
    p += 1
  else :
    f += 1

print(f"합격한 과목수 : {p}")
print(f'불합격한 과목수 : {f}')
```

###4. 학생수와 학생별 점수를 입력받아 리스트에 저장하고 각 점수별 순위를 출력하시오

```python
score=[]

cnt=int(input('학생 수 : '))

for i in range(cnt) :
  s = int(input(f'{i+1}번째 학생점수 : '))
  score.append(s)

rank=[0]*cnt
n=len(score)
for i in range(n) :
  rank[i] = 1
  for j in range(n) :
    if score[i] < score[j] :
      rank[i] += 1
print('---------  학생별 등수 ---------------')
for i in range(n) :
  print(f"{i+1}번째 학생의 점수 : {score[i]} 순위 : {rank[i]}등")
```

###5. 아래 메뉴를 dictionary에 저장하고 출력결과와 같이 출력하고 검색기능을 추가하세요

딕션어리.PNG

```python
menu={}

menu['라면'] = 3000
menu['떡볶기'] = 4000
menu['김밥'] = 2000
menu['햄버거'] = 5000

for i in menu.keys() :
  print(f"{i}-{menu[i]}원")

while True :
  s=input("찾고자 하는 매뉴 (종료:q)")
  if s=='q' :
    print("프로그램 종료")
    break
  if s in menu.keys() :
    print(f"{s}는 {menu[s]}원입니다.")
  else :
    print(f"{s}는 메뉴에 없습니다.")
```

###6. 스마트폰에서 친구를 등록하고 친구 이름으로 검색한 후 연락처를 출력하는 프로그램을 작성 하시오
###선택메뉴 : 1)친구등록  2)검색 3) 종료

```python
addr={}
while True :
  s = int(input("1)친구 등록 2) 검색 3) 종료 : "))
  if s == 1 :
    name = input("이름 : ")
    phone = input("연락처 : ")
    addr[name] = phone
  elif s == 2 :
    name=input("검색할 이름 : ")
    if name in addr.keys() :
        print(f"{name} : {addr[name]}")
    else :
        print(f"{name}은 없습니다.")
  elif s == 3 :
    print("프로그램 종료")
    break
  else :
    print("잘못된 선택 메뉴 입니다. ")
```
