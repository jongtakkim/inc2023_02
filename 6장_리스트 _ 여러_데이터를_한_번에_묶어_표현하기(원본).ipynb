#### Self Check

```python
# 1번 - 1
#아래 book_name 리스트 변수에 책 이름이 저장되어 있다
# book_price 리스트에 책의 가격을 정하는데 15000원~25000원까지 랜덤으로 가격을 정하시오
import random

book_name=['역사탐험대','파이썬','학습법','영단어','여행에세이','삼국지']
book_price = []

for i in range(len(book_name)) :
    book_price.append(random.randint(15000, 25000))

print(book_name)
print(book_price)
```

```python
# 1번 - 2
# 위 문제에서 가장 비싼 책의 이름은 무엇인지 출력하시오
expensive = max(book_price)
idx = book_price.index(expensive)
print('가장 비싼 책은', book_name[idx])
```

```python
# 2번 -1)
# 오디션 참가자 순위표
# 순위 참가자   이번주 득표수   지난주 득표수
#  1   라이언   956412          564553
#  2   니니즈   861832          927175
#  3   어피치   796354          593175
#  4   프로도   534840          864326
#  5   네오     387896          648367
# 순위-이름을 쌍으로 출력하시오
ranking = [[1, '라이언', 956412, 564553],
           [2, '니니즈', 861832, 927175],
           [3, '어피치', 796354, 593175],
           [4, '프로도', 534840, 864326],
           [5, '네오', 387896, 648367]]

for row in ranking :
    print(row[:2])
```

```python
## 2번 - 2)
# 이번주 득표수가 50만 표가 넘는 참가자의 이름과 이번주 득표수를 출력하시오
for row in ranking :
    if row[2] > 500000 :
        print(row[1:3])
```

```python
# 2번 - 3)
# 지나주보드 득표수가 떨어진 참가자가 누구인지 득표수가 얼마나 떨어졌는지를 출력하시오
for row in ranking :
    if row[-2] < row[-1] :
        print(row[1], row[-1]-row[-2], '표 하락')  # 또는 print(row[1], row[-2]-row[-1], '표 하락')
```
